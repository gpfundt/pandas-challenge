```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "../Instructions/Resources/schools_complete.csv"
student_data_to_load = "../Instructions/Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
num_schools = len(school_data)
num_students = len(student_data)
```

```python
total_budget = school_data["budget"].sum()
```

```python
avg_math = student_data["math_score"].mean()
avg_read = student_data["reading_score"].mean()
score = (avg_math + avg_read)/2

pass_math = len(student_data.loc[student_data["math_score"]>=70,:])
percent_pass_math = pass_math/num_students * 100

pass_read = len(student_data.loc[student_data["reading_score"]>=70,:])
percent_pass_read = pass_read/num_students * 100

pass_rate = (percent_pass_math + percent_pass_read) / 2

district_df = pd.DataFrame([{"Total Schools":num_schools,"Total Students":num_students,"Total Budget":total_budget.format("$",
                            "Average Math Score":avg_math, "Average Reading Score":avg_read, "Average Score":score,
                            "% Passing Math":percent_pass_math,"% Passing Reading":percent_pass_read}])
district_df
 
```
